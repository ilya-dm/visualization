```python
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sns
rcParams['figure.figsize'] = 12,9
```

Топ 10 колледжей по количеству игравших в NBA студентов

```python
df = pd.read_csv('nba-players-stats\Players.csv')
df = df.groupby(['collage']).size().reset_index(name='count')
df = df.sort_values('count', ascending = False).head(10)
plt.barh(df['collage'], df['count'])
```

Среднее время проведенное игроком в поле за сезон

```python
df = pd.read_csv('nba-players-stats\Seasons_Stats.csv')
df = df[['Player', 'MP', 'Year']].groupby(['Year']).mean()
df.plot()
```

```python
df = pd.read_csv(r'tmdb-movie-metadata\tmdb_5000_movies.csv')
# df = df.sort_values('passenger_count', ascending=False)
rating = df.vote_average.hist(bins =40)
rating.set_title('User Score distribution')
rating.set_xlabel('User score')
rating.set_ylabel('Films')
```

```python
df = pd.read_csv(r'tmdb-movie-metadata\tmdb_5000_movies.csv')
top_languages = df.original_language.value_counts().sort_values(ascending = False).head(10).index.values
top_languages
sns.boxplot(x="original_language", y="vote_average", 
             data=df[df.original_language.isin(top_languages)])
```
